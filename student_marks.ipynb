```python
import pandas as pd
import numpy as np
import matplotlib .pyplot as plt

#load dataset
df = pd.read_csv("student_marks.csv")

#replace missing marks with column mean
df.fillna(df.mean(numeric_only=True), inplace=True)

#subject wise analysis
Subject= ["Python","SQL","PowerBI"]
 

print("\n ---subject wise average")
print(df[Subject].mean())

print("\n ---subject wise highest marks")
print(df[Subject].max())

print("\n ---subject wise lowest marks")
print(df[Subject].min())

print("\n ---subject wise standard daviation---")
print(df[Subject].std())

print("\n ---subject wise average")
print(df[Subject].mean())


# add total, percentage and grade 
df["Total"]=df[Subject].sum(axis=1)
df["Percentage"]=df["Total"]/(len(Subject)*100)*100

def assign_grade(p):
    if p>=85:
        return "A"
    elif p>=70:
        return "B"
    elif p>=55:
        return "C"
    else:
        return "D"
df["Grade"]=df["Percentage"].apply(assign_grade)
print("---updated data with total,percentage,grade---")
print(df.head())


#filters\conditons
#top 5 students
top_students=df.sort_values("Total",ascending=False).head(5)
```
